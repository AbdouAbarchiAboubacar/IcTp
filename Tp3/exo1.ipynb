```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X_train = np.array([3,4,4,17,24,45,55,68,73])
y_train = np.array([0.9,1.3,1,2.4,2.8,4.4,5.2,6.3,6.6])
```

```python
plt.scatter(X_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
X_train = X_train.reshape(9,1)
y_train = y_train.reshape(9,1)
model.fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
predictions = model.predict(X_train)
plt.scatter(X_train,y_train)
plt.plot(X_train,predictions,c='r')
```

```python
newPred = np.array([28]).reshape(1,1)
model.predict(newPred)
```

```python
from sklearn.metrics import *
```

```python
mse = mean_squared_error(X_train,predictions)
mae = mean_absolute_error(X_train,predictions)
mdae = median_absolute_error(X_train,predictions)
mdae
```

```python
err = np.square(X_train-predictions)
plt.hist(err,bins=50)
```

```python
from scipy.stats import norm
X = np.array([-0.1692 ,-0.149,-0.0524,-0.0192,-0.0168,0.0436 ,0.05572,0.151,0.214])
plt.plot(X, norm.pdf(X,0,1))
plt.show()
```

```python
mean = 0; std = 1; variance = np.square(std)
x = np.array([-0.1692 ,-0.149,-0.0524,-0.0192,-0.0168,0.0436 ,0.05572,0.151,0.214])
f = np.exp(-np.square(x-mean)/2*variance)/(np.sqrt(2*np.pi*variance))

plt.plot(x,f)
plt.ylabel('gaussian distribution')
plt.show()
```

```python
import scipy.stats as stats
import math
XX = np.array([-0.1692,0.151,-0.149,0.214,0.05572,-0.0192,-0.0168,0.0436,-0.0524])
mu = 0
variance = 1
sigma = math.sqrt(variance)
XX = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(XX, stats.norm.pdf(XX, mu, sigma))
plt.show()
```
